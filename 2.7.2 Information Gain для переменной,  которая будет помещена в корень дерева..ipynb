```python
from sklearn import tree
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("E:\\pyrhon\\progectTEST\\2.7\\train_data_tree.csv")
```

```python
X_test = data.drop(['num'], axis=1)
y_test = data['num']
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
clf.fit(X_test,y_test)
```

```python
tree.plot_tree(clf, feature_names=X_test.columns)
```

Из дерева можно узнать интересующую нас энтропию и кол-во образцов(samples).
В нашем случае N = 238.
n1 = 157, n2 = 81
E1 = 0,903, E2 = 0.826
Теперь можем посчитать IG(Information Gain):
IG = E(Y) - E(Y|X)
E(Y) = 0.996, а E(Y|X) = (n0*E0 + n1*E1)/N тогда 
IG = 0.996 -((157÷238)×0,903+(81÷238)×0,826) = 0.119
