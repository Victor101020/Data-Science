### Формула изменения весов стохастического градиентного спуска

$w = w - α\frac{δloss}{δw},\\$
где $\frac{δloss}{δw} = 2 * x_n * (x_n * w - y_n)$

### Функция потерь стохастического градиентного спуска
$loss = (\hat y - y)^2 = (x*w - y)^2$

### Запрограммируем алгоритм стохастического градиентного спуска

```python
import matplotlib.pyplot as plt
```

```python
# Подготовим данные для обучения
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
```

```python
# Начальный вес
w = 1.0
```

```python
# Определите модель 
def forward(x):
  return x * w
```

```python
# Определить функцию потерь
def loss(x, y):
  y_pred = forward(x)
  return (y_pred - y) ** 2
```

```python
# Определите функцию стохастического градиентного спуска
def gradient(x, y):
  return 2 * x * (x * w - y)
```

```python
# Распечатайте градиент перед тренировкой
print('Predict (before training)', 4, forward(4))
```

```python
# Тренировка 
loss_list = []
epoch_list = []
for epoch in range(100):
  for x, y in zip(x_data, y_data):
    grad = gradient(x, y)
    w -= 0.01 * grad
    print("\tgrad:", x, y, grad)
    l = loss(x, y)
  print("epoch:", epoch, "w=", w, "loss=", loss)
  epoch_list.append(epoch)
  loss_list.append(l)
print('Predict(after training)', 4, forward(4))
```

```python
# Визуализируем, как тренировалась наша модель
plt.plot(epoch_list, loss_list)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```
