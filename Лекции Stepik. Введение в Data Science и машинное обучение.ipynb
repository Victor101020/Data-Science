2 ЧАСТЬ КУРСА. РЕШАЮЩИЕ ДЕРЕВЬЯ

```python
data = pd.DataFrame({'X_1': [1, 1, 1, 0, 0, 0, 0, 1], 'X_2': [0, 0, 0, 1, 0, 0, 0, 1], 'Y': [1, 1, 1, 1, 0, 0, 0, 0]})
```

Наша задача заключается в предсказании ,при помощи деревьев решений,  у на основании х1 и х2

```python
clf = tree.DecisionTreeClassifier(criterion = 'entropy') #создаем дерево решений с помощью tree
```

Требуется преобразовать данные: Сохраним все фичи в переменную Х, а целевую переменную сохраним в y

```python
X = data[['X_1','X_2']]
y = data.Y
```

Все готово для первого обучения первого дерева решений. Воспользуемся методом  fit()

```python
clf.fit(X, y)
```

Визуализируем дерево решений

```python
from IPython.display import SVG
from graphviz import Source
from IPython.display import display

from IPython.display import HTML
style = "<style>svg{width:70% !important;height:70% !important;}</style>"
HTML(style)
```

```python
clf.fit(X, y)

tree.plot_tree(clf, feature_names=list(X),
               class_names=['Negative', 'Positive'],
               filled=True)
```

```python
graph = Source(tree.export_graphviz(clf, out_file=None,
                                   feature_names=list(X),
                                   class_names=['Negative','Positive'],
                                   filled = True))

display(SVG(graph.pipe(format='svg')))
```
