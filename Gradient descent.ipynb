```python
bounds = np.array([[-10.0,10.0]])
```

```python
x = bounds[:, 0] +rand(len(bounds))*(bounds[: ,1] - bounds[:,0])
```

```python
x
```

```python
inputs = []
outputs = []
```

```python
def objective(x):
    return x**2.0
```

```python
for i in range(1000):
    x = bounds[:,0] +rand(len(bounds))*(bounds[: ,1] - bounds[:,0])
    y = objective(x)
    inputs.append(x)
    outputs.append(y)
    
```

```python
plt.figure(figsize = [20.0,10.0])
plt.scatter(inputs,outputs)
plt.ylabel('Значение функции ошибки, E', fontsize=16)
plt.xlabel('Вес модели, w', fontsize=16)
```

```python
def derevative(x):
    return x*2.0
```

```python
def objective(x):
    return x**2.0
```

```python
solutions = []
evaluations = []
```

```python
x = -10.0

for i in range(20):
    solutions.append(x)
    x_evaluation = objective(x)
    evaluations.append(x_evaluation)
    
    gradient = derevative(x)
    new_x = x - 0.1 * gradient
    x = new_x
    print('Шаг алгоритма {}, значение веса {}, функция ошибки {}'.format(i,x,x_evaluation))
fig = plt.figure(figsize = [20.0,10.0])
plt.scatter(inputs, outputs)
plt.scatter(solutions, evaluations, color = 'red')
plt.ylabel('Значение функции ошибки, E', fontsize=16)
plt.xlabel('Вес модели, w', fontsize=16)
```
