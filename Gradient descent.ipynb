## Импортируем необходимые библиотеки

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import asarray
from numpy import arange
from numpy.random import rand
import matplotlib.animation as animation
from celluloid import Camera
```

## Генерируем списки значений x и y для построения зависимости

```python
bounds = np.array([[-10.0,10.0]])
```

```python
inputs = []
outputs = []
```

```python
def objective(x):
    return x**2.0
```

```python
for i in range(1000):
    x = bounds[:,0] +rand(len(bounds))*(bounds[: ,1] - bounds[:,0])
    y = objective(x)
    inputs.append(x)
    outputs.append(y)
    
```

## Строим график зависимости функции ошибки от веса модели

```python
plt.figure(figsize = [20.0,10.0])
plt.scatter(inputs,outputs)
plt.ylabel('Значение функции ошибки, E', fontsize=16)
plt.xlabel('Вес модели, w', fontsize=16)
```

## Реализовывываем градиентный спуск

```python
# функция производной
def derevative(x):
    return x*2.0
```

```python
# функция ошибки
def objective(x):
    return x**2.0
```

```python
solutions = []
evaluations = []
x = -10.0

for i in range(20): # запускаем цикл обучения
    solutions.append(x) # добавляем значение x  в список, чтобы потом построить точки на графике
    x_evaluation = objective(x) # вычисляем значение функции ошибки с этим весом
    evaluations.append(x_evaluation) # добавляем в список значение функции ошибки
    
    gradient = derevative(x) # вычисляем производную по нынешнему весу
    new_x = x - 0.1 * gradient # изменяем значение веса(улучшаем вес)
    x = new_x # присваиваем значение нового веса
    print('Шаг алгоритма {}, значение веса {}, функция ошибки {}'.format(i,x,x_evaluation)) # печать в консоль

# строим финальный график
fig = plt.figure(figsize = [20.0,10.0])
plt.text(-10.1, 0.2, "шаг градиента = 0.1", fontsize = 18)
plt.scatter(inputs, outputs)
plt.scatter(solutions, evaluations, color = 'red')
plt.ylabel('Значение функции ошибки, E', fontsize=16)
plt.xlabel('Вес модели, w', fontsize=16)
```
